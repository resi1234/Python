```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
page=requests.get("http://scores.sify.com")
page


# In[47]:


#page.content


# In[48]:


soup=BeautifulSoup(page.content,'html.parser')
```

```python
url=[]
count_url=0
for i in soup.find_all('div',class_="scoresbox-center"):
    try:
        url.append(i.h2.a.get('href'))
        count_url+=1
    except:
        break
print(url)    
print(count_url)        
```

```python
score=[]
det=[]

for i in range(count_url):
    score.append(soup.find('h3',id='batteamnameruns'+str(i+1)).text)
    try:
        det.append(soup.find('h4',id='batsmanbowlerdetails'+str(i+1)).text)
    except:
        pass
det   
    
```

```python
c=0
for i in url:
    page=requests.get(i)
    soup=BeautifulSoup(page.content,'html.parser')
    
    
    #briefing about the match
    i=0
    for a in soup.find('div',class_='team-live').strings:
        if i >0 and i!=4: 
            print("###########################")
        if(i!=4):
            print(a)
        i+=1
    print('##############################\n\n')
    print(score[c])
    try:
        sp=det[c].split(';')
        print(sp[0]+'    '+sp[1])
        print(sp[2]+'    '+sp[3])
    except:
        pass
    c+=1
    print('\033[1m'+"\nScorecard"+'\033[0m')
    
        
    #extracting the main class which contains scorecard 
    ss=soup.find('div',class_='col-md-6')
    
    
    #contains scoreboard of the batting team 
    my_table=ss.find(class_='table table-hover batting')
    
    #names of the batsman
    bat=[]
    for i in my_table.find_all('a'):
        bat.append(i.text)
    
   
    #for extracting rows from the table body of batting scaorecard
    row=my_table.find('tbody').find_all('tr')
    
    
    #for extracting runs of the batsman
    runs=[]
    for i in row[:-2]:
        col=i.find_all('td')
        counter=0
        for i in col:
            if counter==2:
                runs.append(i.text)
            counter+=1    
    
    
    #balls played by him
    balls=[]
    for i in row[:-2]:
        col=i.find_all('td')
        counter=0
        for i in col:
            if counter==3:
                balls.append(i.text)
            counter+=1
            
    
    
    fours=[]
    for i in row[:-2]:
        col=i.find_all('td')
        counter=0
        for i in col:
            if counter==4:
                fours.append(i.text)
            counter+=1
    
    
    sixes=[]
    for i in row[:-2]:
        col=i.find_all('td')
        counter=0
        for i in col:
            if counter==5:
                sixes.append(i.text)
            counter+=1
    
    
    sr=[]
    for i in row[:-2]:
        col=i.find_all('td')
        counter=0
        for i in col:
            if counter==6:
                sr.append(i.text)
            counter+=1
            
            
    
    
    ScoreBoard=pd.DataFrame({
    "Batsman":bat,
     "Runs":runs,
      "Balls":balls,
       "4s":fours,
        "6s":sixes,
         "Strike_rate":sr
    })
    ScoreBoard=ScoreBoard.set_index("Batsman")
    print(ScoreBoard)
    
    #extracting the bowling scorecard
    try:
        my_table=ss.find_all(class_='table table-hover')[1]
    
    
        #extracting the bowlers name
        bowl=[]
        for i in my_table.find_all('a'):
            bowl.append(i.text)     
    
        row=my_table.find('tbody').find_all('td')
    
    
        #extracting the overs
        overs=[]
        for i in range(1,len(row),7):
            overs.append(row[i].text)
        
    
    
        maidens=[]
        for i in range(2,len(row),7):
            maidens.append(row[i].text)
    
    
        runs=[]
        for i in range(3,len(row),7):
            runs.append(row[i].text)
    
    
    
        wickets=[]
        for i in range(4,len(row),7):
            wickets.append(row[i].text)
    
    
    
        eco=[]
        for i in range(5,len(row),7):
            eco.append(row[i].text)
    
    
        ScoreBoard=pd.DataFrame({
        "Bowler":bowl,
         "Overs":overs,
         "Maidens":maidens,
          "Runs":runs,
           "Wickets":wickets,
            "Economy":eco
        })
        ScoreBoard=ScoreBoard.set_index("Bowler")
        print(ScoreBoard)
    except:
        pass
    
    
    try:
        bat=[]
        inng_2_bat=soup.find_all('tbody')[3]
        for i in inng_2_bat.find_all('a'):
            bat.append(i.text)
    
    
        row=inng_2_bat.find_all('tr')
    
    
        runs=[]
        for i in row[:-2]:
            col=i.find_all('td')
            runs.append(col[2].text)
        
    
        balls=[]
        for i in row[:-2]:
            col=i.find_all('td')
            balls.append(col[3].text)
    
        fours=[]
        for i in row[:-2]:
            col=i.find_all('td')
            fours.append(col[4].text)
    
    
        sixes=[]
        for i in row[:-2]:
            col=i.find_all('td')
            sixes.append(col[5].text)
    
    
        sr=[]
        for i in row[:-2]:
            col=i.find_all('td')
            sr.append(col[6].text)
        
    
        ScoreBoard=pd.DataFrame({
        "Batsman":bat,
         "Runs":runs,
          "Balls":balls,
           "4s":fours,
            "6s":sixes,
             "Strike_rate":sr
        })
        ScoreBoard=ScoreBoard.set_index("Batsman")
        print(ScoreBoard)
    except:
        pass
    
    
    try:
        row=soup.find_all('tbody')[5].find_all('td')
        
        
        overs=[]
        for i in range(1,len(row),7):
            overs.append(row[i].text)
        maidens=[]
        for i in range(2,len(row),7):
            maidens.append(row[i].text)
        runs=[]
        for i in range(3,len(row),7):
            runs.append(row[i].text) 
        wickets=[]
        for i in range(4,len(row),7):
            wickets.append(row[i].text)
        eco=[]
        for i in range(5,len(row),7):
            eco.append(row[i].text)
        bowl=[]
        for i in row:
            try:
                for j in i.find_all('a'):
                    bowl.append(j.text)
            except:
                pass
        
        ScoreBoard=pd.DataFrame({
            "Bowler":bowl,
             "Overs":overs,
             "Maidens":maidens,
              "Runs":runs,
               "Wickets":wickets,
                "Economy":eco
            })
        ScoreBoard=ScoreBoard.set_index("Bowler")
        print(ScoreBoard)
    
    except:
        pass
    print('$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
    print('$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
    print('                    $$$$$$$$$$$$$$$$$$$$$$$$$$$')
    print('$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
    print('$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
    
```
